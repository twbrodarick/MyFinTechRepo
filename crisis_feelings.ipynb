### Calculating VADER Sentiment Score for News Titles and Text

As you know the `compound` score could be used to get a normalized score for a sentiment, in this section you have to create a function called `get_sentiment(score)` that will return a normalized value of sentiment for the `score` parameter based on the rules you learn. This function should return `1` for positive sentiment, `-1` for negative sentiment, and `0` for neutral sentiment.

```python
# Sentiment calculation based on compound score
def get_sentiment(score):
    """
    Calculates the sentiment based on the compound score.
    """
    if score >= 0.05:
        return 1
    elif score <= -0.05:
        return -1
    else:
        return 0
```

Use the the VADER sentiment module from `NLTK` to score the sentiment of every news article title and text in english; you should append ten new columns to the English news DataFrame to store the results as follows.

* Title's compound score
* Title's positive score
* Title's neutral score
* Title's negative score
* Title's normalized score (using the `get_sentiment()` function)
* Text's compound score
* Text's positive score
* Text's neutral score
* Text's negative score
* Text's normalized score (using the `get_sentiment()` function)

```python
def ScoreSentimentAll(row):
    try:
        text = row["title"]
        sentiment = analyzer.polarity_scores(text)
        return sentiment
    except AttributeError:
        pass
    
def ScoreSentimentTC(row):
    try:
        sentiment = row["title_sentiment"]
        return sentiment["compound"]
    except AttributeError:
        pass
    
def ScoreSentimentTP(row):
    try:
        sentiment = row["title_sentiment"]
        return sentiment["pos"]
    except AttributeError:
        pass
    
def ScoreSentimentTNeg(row):
    try:
        sentiment = row["title_sentiment"]
        return sentiment["neg"]
    except AttributeError:
        pass
    
def ScoreSentimentTN(row):
    try:
        sentiment = row["title_sentiment"]
        return get_sentiment(sentiment["compound"])
    except AttributeError:
        pass

def ScoreSentimentXAll(row):
    try:
        text = row["text"]
        sentiment = analyzer.polarity_scores(text)
        return sentiment
    except AttributeError:
        pass    
    
def ScoreSentimentXC(row):
    try:
        sentiment = row["text_sentiment"]
        if sentiment:
            return sentiment["compound"]
    except AttributeError:
        pass
    
def ScoreSentimentXP(row):
    try:
        sentiment = row["text_sentiment"]
        if sentiment:
            return sentiment["pos"]
    except AttributeError:
        pass
    
def ScoreSentimentXN(row):
    try:
        sentiment = row["text_sentiment"]
        if sentiment:
            return get_sentiment(sentiment["compound"])
    except AttributeError:
        pass

df["title_sentiment"] = df.apply(ScoreSentimentAll, axis=1)
df["title_compound"] = df.apply(ScoreSentimentTC, axis=1)
df["title_pos"] = df.apply(ScoreSentimentTP, axis=1)
df["title_neg"] = df.apply(ScoreSentimentTNeg, axis=1)
df["title_normal"] = df.apply(ScoreSentimentTN, axis=1)
df["text_sentiment"] = df.apply(ScoreSentimentXAll, axis=1)
df["text_compound"] = df.apply(ScoreSentimentXC, axis=1)
df["text_pos"] = df.apply(ScoreSentimentXP, axis=1)
df["text_normal"] = df.apply(ScoreSentimentXN, axis=1)
df
```
